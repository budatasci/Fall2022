```python
data_num = data[['Age', 'Tumor Size', 'Reginol Node Positive', 'Survival Months', 'Regional Node Examined']]
data_cat = data[['Race', 'Marital Status', 'T Stage ', 'N Stage', '6th Stage', 'differentiate', 'Grade', 'A Stage','Estrogen Status',
                  'Progesterone Status', 'Status']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for i in data_cat.columns:
    sns.barplot(data_cat[i].value_counts().index,data_cat[i].value_counts()).set_title(i)
    plt.show()
```

```python
for i in data_num.columns:
    plt.hist(data_num[i])
    plt.title(i)
    plt.show()
```

```python
for col in data_cat.columns:
  df = data_cat.groupby([col, 'Status'])[col].count().unstack('Status')
  df[['Alive','Dead']].plot(kind='bar', stacked=True)
```

```python
sns.heatmap(data_num.corr())
```

```python
for col in data_num.columns:
  sns.boxplot(x = col, y = 'Status', data = data)
  plt.show()
```
